### 생각의 흐름
#### state를 받아서 oroginal position과 current position을 추출
#### poisition index로 색의 나열을 찾는다
#### 찾을 수 있는 이유는 색의 조합은 유일하고
#### target face의 색상은 유지되고 여러 색의 나열 속에서 색의 조합과 일치하는 것을 찾고
#### 그 중 position index에 있는 색이 초기 target face의 색상과 동일한 것을 찾으면 되기 때문이다
#### 그러려면 모든 경우의 수를 포함하는 reference가 정확해야 한다

```python
base_cube = pc.Cube()
corner_location_list = ["LDB", "LDF", "LUB", "LUF", "RDB", "RDF", "RUB", "RUF"]
for origin_position, state_position in enumerate(corner_state):
    state_position = np.where(state_position == 1.0)[0][0] # index of case
    position, position_idx = state_position // 3, state_position % 3 # 실제 위치는 position, 원래 위치를 알기 위해서는 color set이 필요
    current_location = corner_location_list[position]
    origin_location = corner_location_list[origin_position]
    # 그 코너의 원래 색상
    base_color_map = base_cube[origin_location].facings

    base_set = set()
    for s in origin_location:
        base_set.add(base_color_map[s])
    # 기준 면의 색 얻기 -> 기준 면의 색이 정해지면 나머지 두 면의 색도 고정 왜냐하면 xyz축의 관계가 고정되어 있기 때문

    # for count, base_key in enumerate(base_color_map.keys()):
    #     if count == position_idx:
    #         target_color = base_color_map[base_key] # Square type

    # 색 조합은 유일
    # 사용하는 기준 면은 각 코너 큐브의 U 와 D 
    # target color가 position idx에 있다 -> 그 때 나머지 두 색은 어떻게 나열되는가? -> 정답지가 있다
    # 각 코너 큐브는 코너에 위치할 때 색의 xyz 조합이 정해져 있다
    # 그 조합에 target color와 위치를 넣으면 나머지 두 색의 조합도 알 수 있다
    # 그 조합을 Corner( ) 에 넣으면 Corner class 가 생성된다
    reference_facings_list = reference[origin_location][current_location] # 중복이 없는 코너 큐브의 가능한 모든 경우
    # 코너 큐브 이름 : {코너 큐브의 현재 위치: facings list}
    for reference_facings in reference_facings_list:
        refer_set = set()
        for s in reference_facings.keys():
            refer_set.add(reference_facings[s])

        # if base_color_map == reference_facings: # 색 조합이 같은 것을 찾는다, 색 조합은 같지만 순서가 다르다 reference_facings가 실제 색상이다
        if base_set == refer_set:
            key_list = []
            for count, refer_key in enumerate(reference_facings.keys()):
                key_list.append(refer_key)
                if count == position_idx:
                    target_color = reference_facings[refer_key]
                    target_key = refer_key
            # 이를 바탕으로 새로운 코너 큐브를 만든다 그러기 위해서는 key string이 필요하다 corner cube의 정의를 위해
            for idx, corner_name in enumerate(corner_location_list):
                # done = False
                if (key_list[0] in corner_name) and (key_list[1] in corner_name) and (key_list[2] in corner_name):
                    idx_list = [corner_name.index(key_list[0]), corner_name.index(key_list[1]), corner_name.index(key_list[2])]
                    idx_list.sort()
                    # idx_0 = corner_name.index(key_list[0]) 
                    # idx_1 = corner_name.index(key_list[1]) 
                    # idx_2 = corner_name.index(key_list[2]) 
                    # done = True
                # if done:
                    if idx == 0:
                        cubie_set.add(
                        Corner(
                            L = reference_facings[corner_name[idx_list[0]]],
                            D = reference_facings[corner_name[idx_list[1]]],
                            B = reference_facings[corner_name[idx_list[2]]]
                            ))
                    elif idx == 1:
                        cubie_set.add(
                        Corner(
                            L = reference_facings[corner_name[idx_list[0]]],
                            D = reference_facings[corner_name[idx_list[1]]],
                            F = reference_facings[corner_name[idx_list[2]]]
                            ))
                    elif idx == 2:
                        cubie_set.add(
                        Corner(
                            L = reference_facings[corner_name[idx_list[0]]],
                            U = reference_facings[corner_name[idx_list[1]]],
                            B = reference_facings[corner_name[idx_list[2]]]
                            ))
                    elif idx == 3:
                        cubie_set.add(
                        Corner(
                            L = reference_facings[corner_name[idx_list[0]]],
                            U = reference_facings[corner_name[idx_list[1]]],
                            F = reference_facings[corner_name[idx_list[2]]]
                            ))
                    elif idx == 4:
                        cubie_set.add(
                        Corner(
                            R = reference_facings[corner_name[idx_list[0]]],
                            D = reference_facings[corner_name[idx_list[1]]],
                            B = reference_facings[corner_name[idx_list[2]]]
                            ))
                    elif idx == 5:
                        cubie_set.add(
                        Corner(
                            R = reference_facings[corner_name[idx_list[0]]],
                            D = reference_facings[corner_name[idx_list[1]]],
                            F = reference_facings[corner_name[idx_list[2]]]
                            ))
                    elif idx == 6:
                        cubie_set.add(
                        Corner(
                            R = reference_facings[corner_name[idx_list[0]]],
                            U = reference_facings[corner_name[idx_list[1]]],
                            B = reference_facings[corner_name[idx_list[2]]]
                            ))
                    elif idx == 7:
                        cubie_set.add(
                        Corner(
                            R = reference_facings[corner_name[idx_list[0]]],
                            U = reference_facings[corner_name[idx_list[1]]],
                            F = reference_facings[corner_name[idx_list[2]]]
                            ))
                    else:
                        pass


            
```
